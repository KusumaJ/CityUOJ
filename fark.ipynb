```python
def f(f):
    if(f==0): return 5
    elif(f==1): return 6
    elif(f==2): return 7
    elif(f==3): return 1
    elif(f==4): return 8
    elif(f==5): return 3
    else: return 2

def function2(n):
    if(n<=7): return f(n)
    else: 
        # print(n-1,n-2,n-3)
        # print(f(n-1),f(n-2),f(n-3))
        return function2(n-1)+function2(n-2)-function2(n-3)


for i in range(8,201):
    print(i)
    print("***")
    print(function2(i))
    print("***")
# print(function2(8))
```
